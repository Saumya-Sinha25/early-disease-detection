```python
import pandas as pd

# Load the raw diabetes dataset (stored in the same folder as this notebook)
df = pd.read_csv("diabetes.csv")

# Show basic info
print(df.info())
print(df.describe())
```

```python
import numpy as np

# Columns where 0 is not a valid value
invalid_zero_cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Replace 0 with NaN in these columns
df[invalid_zero_cols] = df[invalid_zero_cols].replace(0, np.nan)

# Check how many missing values are introduced
print(df.isnull().sum())
```

```python
# Fill missing values with median of each column
for col in invalid_zero_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Confirm no more missing values
print(df.isnull().sum())
```

```python
%pip install scikit-learn
from sklearn.preprocessing import StandardScaler

# Don't scale the 'Outcome' column
features_to_scale = df.columns.drop('Outcome')

# Initialize the scaler
scaler = StandardScaler()

# Apply scaling
df[features_to_scale] = scaler.fit_transform(df[features_to_scale])

# Preview scaled data
df.head()
```

```python
# Save the cleaned dataset in the same folder as this notebook
df.to_csv("diabetes_cleaned.csv", index=False)
print("Cleaned dataset saved successfully!")
# Display the first few rows of the cleaned dataset
print(df.head())
```
